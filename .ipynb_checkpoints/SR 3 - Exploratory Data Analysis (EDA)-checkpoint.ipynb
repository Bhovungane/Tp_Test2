```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\27797\Desktop\TP2_Test2\Informal_Settlement_Programme_-1898186875815614491.csv")
```

```python
# 1. Basic Infographic Charts

# Bar chart: Population Density by Region
plt.figure(figsize=(10, 6))
sns.barplot(x='REGION', y='LATITUDE', data=df, ci=None)
plt.title(' LATITUDE')
plt.xlabel('REGION')
plt.ylabel('LATITUDE')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram: Distribution of Population Density
plt.figure(figsize=(10, 6))
sns.histplot(df['LATITUDE'], bins=20, kde=True)
plt.title('Distribution of LATITUDE')
plt.xlabel('LATITUDE (people per sq km)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count the occurrences of each sanitation type
sanitation_counts = df['SANITATION'].loc[100:109].value_counts()

# Plot a pie chart for sanitation types
plt.figure(figsize=(8, 8))
plt.pie(sanitation_counts, labels=sanitation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Sanitation Types Distribution')
plt.show()
```

```python
# Example: Plotting 'Population_Density' by 'Region' using a violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='REGION', y='LATITUDE', data=df)
plt.title('Population Density Distribution by Region (Violin Plot)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select numeric columns for pairplot
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Plot pairplot
sns.pairplot(df[numeric_columns])
plt.title('Pairplot of Numeric Features')
plt.show()
```
